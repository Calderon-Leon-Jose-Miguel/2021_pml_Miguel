```python
values = [5,52,45,3,14,-45,6,-90,-76,-2,-79] # damos una data
smallest, largest = None, None # definoms smalles y largest y le damod el valor None 
for v in values:
    if smallest == None and largest == None: # si los valores valen Nome (la primera interacion), asignamos el valor v a samllest y largest
        smallest, largest = v, v
    else: #todas las interacciones que no sean la primera entran en este else
        smallest = min(smallest, v) # vamos comparando el smallest con el valor v y tomamos el menor 
        largest = max(largest, v)  # vamos comparando el largest con el valor v y tomamos el meyor 
print(smallest, largest)
#las desventajas es que tiene complejidad n^2, es decir la cantidad de tiempo  que necesita  para trabjar con datas gigantes sería  mucho
#las ventajas el que el codigo es muy intuitivo. 
```

```python
def get_egg_label(mass):
    egg_label = "Unlabelled" #definismos una variable local,  la cual será el str Unlabelled"
    #copiamos el codigo dado 
    if mass >= 90:
        egg_label = "warning: egg might be dirty"
    elif mass >= 85:
        egg_label = "jumbo"
    elif mass >= 70:
        egg_label = "large"
    elif mass < 70 and mass >= 55:
        egg_label = "medium"
    elif mass < 50:
        egg_label = "too light, probably spoiled"
    else:
        egg_label = "small"
    return egg_label

    
```
